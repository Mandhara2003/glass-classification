```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("/content/drive/MyDrive/glass.csv")
```

```python
df = pd.DataFrame(data)

x = df[['RI','Na','Mg','Al','Si','K']]
y = df['Type']
```

```python
# Split data into training and testing sets (optional, for evaluation)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=3)  # Adjust k as needed
knn.fit(X_train, y_train)  # Train on training data (if using split)
```

```python
# Make predictions on unseen data (replace with your new data)
new_data = np.array([[5.1, 13.64, 3.96, 1.16, 72.73, 0.15]])  # Example data (adjust format as needed)
predictions = knn.predict(new_data)
```

```python
# Print the predicted glass type(s)
for prediction in predictions:
    # Assuming class labels correspond to process types (float = 1, non-float = 2)
    if prediction == 1:
        print("Predicted: Float processed glass")
    else:
        print("Predicted: Non-float processed glass")
```
